```python
path_dic = "C:/Users/ADMIN/Desktop/Thay_Lam/CCCD/imgs/card/test/"

path_img = "C:/Users/ADMIN/Desktop/Thay_Lam/CCCD/imgs/truoc/"
```

```python
data = pd.read_csv(r"C:\Users\ADMIN\Desktop\Thay_Lam\CCCD\codes\data_edit1.csv", header = None)
for i in range(len(data)):
    print(data[0][i])
#     print(type(data[1][i]))
    name = data[0][i].split('\\')[-1]
    print(name)
    img = cv2.imread(path_img + name)
    img = cv2.resize(img, (750, 750))
    h, w, _ = img.shape
    
    tl = [int(data[1][i]* w) , int(data[2][i]* h)]
    tr = [int(data[3][i]* w) , int(data[4][i]* h)]
    br = [int(data[5][i]* w) , int(data[6][i]* h)]
    bl = [int(data[7][i]* w) , int(data[8][i]* h)]
    
    input_pts = np.float32([tl, tr, br, bl])
    output_pts = np.float32([[0, 0], [500, 0], [500, 300], [0, 300]])
    M = cv2.getPerspectiveTransform(input_pts,output_pts)
    out1 = cv2.warpPerspective(img.copy(),M,(500, 300),flags=cv2.INTER_LINEAR)
    cv2.imwrite(path_dic + name, out1)

data.head()
```
